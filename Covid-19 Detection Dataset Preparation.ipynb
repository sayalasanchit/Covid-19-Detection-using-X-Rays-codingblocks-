```python
import pandas as pd
import os
import shutil
```

### Covid Data

```python
file_path="covid-chestxray-dataset-master/metadata.csv"
images_path="covid-chestxray-dataset-master/images"
```

```python
df=pd.read_csv(file_path)
print(df.shape)
df.head(10)
```

```python
if not os.path.exists("Dataset"):
    os.mkdir("Dataset")
```

```python
target_dir="Dataset/Covid"
if not os.path.exists(target_dir):
    os.mkdir(target_dir)
```

```python
count=0
for (i, row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(images_path, filename)
        image_copy_path=os.path.join(target_dir, filename)
        shutil.copy2(image_path, image_copy_path)
        print(f"Copied image {count+1}")
        count+=1
print("Total images copied:", count)
```

### Normal data

```python
import random
file_path="archive/chest_xray/train/NORMAL"
target_dir="Dataset/Normal"
```

```python
if not os.path.exists(target_dir):
    os.mkdir(target_dir)
```

```python
image_names=os.listdir(file_path)
random.shuffle(image_names) # Randomizing the names
# Selecting first 196 images
for i in range(count):
    image_name=image_names[i]
    image_path=os.path.join(file_path, image_name)
    target_path=os.path.join(target_dir, image_name)
    shutil.copy2(image_path, target_path)
    print(f"Copied image {i+1}")
```
